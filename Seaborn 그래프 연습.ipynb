## Matplotlib에서 서브 플롯 크기 설정하기

* gridspec_kw 방법
    - fig, as = plt.subplot( 행, 열, gridspec_kw = dict (width_ratios = [a, b], heigh_ratios = [c, d]))
    - width_ratios는 너비의 비율을 나타낸다. width_ratios = [a, b]는 a:b 이다.
    - heigh_ratios는 높이의 비율을 나타낸다. width_ratios = [c, d]는 c:d 이다.

```python
fig, ax = plt.subplots(1, 2, figsize = (8,4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=df, x='flipper_length_mm', y = 'bill_length_mm', hue='species', ax=ax[0])
sns.histplot(data=df, x='species', hue = 'species', shrink = .8, alpha=.6, legend=False, ax=ax[1])
fig.tight_layout()
```

- shrink 파라미터는 막대의 너비를 조정한다.
- legend = False: 범례를 제거하는 옵션이다.
- fig.tight_layout(): 자동으로 그래프간 공백을 정렬한다.(자동 레이아웃 설정)
