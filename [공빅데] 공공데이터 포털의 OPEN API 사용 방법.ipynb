### step1. 라이브러리 import 및 url입력

```python
# 라이브러리 import
import requests
import pprint # 파이썬 데이터 구조를 예쁘게 인쇄하는 기능을 제공
import json
```

```python
# url 입력
url = 'http://api.data.go.kr/openapi/tn_pubr_public_cctv_api?serviceKey=9dNbZ%2FxAAEdZTV3B%2BxJmwyqgjb4W5TrpKYIYzkFemFeFFseRZdnGbJSSHDf5cJpGZNdMLtIfjnAR%2B%2BBJKGvebg%3D%3D&pageNo=1&numOfRows=100&type=json'
```

```python
# url 불러오기
response = requests.get(url)
```

```python
# 데이터 값 출력해보기
contents = response.text
```

```python
contents
```

```python
# 데이터 결과값 예쁘게 출력해주는 코드
pp = pprint.PrettyPrinter(indent=4) # indent: 들여쓰기 지정
print(pp.pprint(contents))
```

### step2. 문자열을 json으로 변경하기
위의 문자열을 json으로 변경한다.

```python
# 문자열을 json으로 변경
json_ob = json.loads(contents) # 문자열을 사전형(json)으로 전환

print(json_ob)
print(type(json_ob))
```

### step3. 필요한 내용만 남기기
위를 보면 데이터의 내용에서 response 키 값의 body 속에 items가 있고, 우리가 원하는 정보가 모두 들어있음을 알 수 있다. 이를 확인 후 아래와 같이 items 내용만 꺼낸다.

```python
# 필요한 내용만 꺼내기
body = json_ob['response']['body']['items']
print(body)
```

### step4. Dataframe으로 변환하기
위에서 작성한 body 변수에 저장된 내용을 pandas를 활용해서 Dataframe으로 만들어 준다.

```python
import pandas as pd
from pandas.io.json import json_normalize

# Dataframe으로 만들기
df = pd.json_normalize(body)
df
```
